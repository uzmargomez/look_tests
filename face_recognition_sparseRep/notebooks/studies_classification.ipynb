```python
import matplotlib.pyplot as plt
from sklearn.metrics import (
    confusion_matrix, 
    classification_report,
    ConfusionMatrixDisplay,
    roc_curve,
    RocCurveDisplay,
)


plt.rcParams['figure.figsize'] = [15, 15]
plt.rcParams.update({'font.size': 16})

accuracy = []

sizes = [8,16,24,32,40,48,64]

for size in sizes:
    file = open(f'data/output{size}.txt', 'r')
    lines = file.readlines()
    
    y_true = []
    y_pred = []    
    
    for l in lines:
        l = l.replace(" ","").replace("*NOTINDB*","Unknown")
        result = l.split("|")
        
        y_true.append(result[0])
        y_pred.append(result[1])
    
    names = list(set(y_true))
    cm = confusion_matrix(y_true, y_pred,labels=names)
    
    disp = ConfusionMatrixDisplay(cm,display_labels=names)
    disp.plot()
    disp.ax_.tick_params(axis='x', rotation=70)
    plt.savefig(f"data/cm_{size}.png")
    plt.show()
    
    
    cr = classification_report(y_true, y_pred,zero_division=0,output_dict=True)
    print(classification_report(y_true, y_pred,zero_division=0))
    accuracy.append(cr['accuracy'])
    
plt.rcParams['figure.figsize'] = [10, 8]
plt.plot(sizes,accuracy,'ro')
plt.ylim(0,1)
plt.xlabel('Image size')
plt.ylabel('Accuracy')
plt.grid()
plt.savefig(f"data/accuracy.png")
plt.show()
```
